```python
import pandas as pd
import pytesseract
from PIL import Image
import cv2
import requests
from io import BytesIO
import numpy as np
import re
from tkinter import Tk, filedialog
```

```python
# Read the dataset from CSV
input_csv = "/Users/udaysinghshekhawat/Downloads/student_resource 3/dataset/sample_test.csv"  # Path to your dataset file
df = pd.read_csv(input_csv)
df
```

```python

# Function to download image and perform OCR
def extract_text_from_image(image_url):
    response = requests.get(image_url)
    img = Image.open(BytesIO(response.content))
    text = pytesseract.image_to_string(img)
    return text
```

```python



# Function to extract entity (e.g., weight, volume) from OCR text
def extract_entity(text, entity_name):
    if entity_name == 'weight':
        match = re.search(r'(\d+\.?\d*)\s*(kg|g|lbs)', text.lower())  # Adjust for other units if needed
        return match.group(0) if match else 'Not found'
    # Add more entity extraction patterns for other entity types (e.g., volume, voltage, etc.)
    # List to store the results
results = []
```

```python



# Iterate over the dataset rows
for index, row in df.iterrows():
    group_id = row['group_id']          # Use 'group_id' column
    image_url = row['image_link']       # Use 'image_link' column
    entity_name = row['entity_name']    # Use 'entity_name' column

    
    # Perform OCR and entity extraction
    ocr_text = extract_text_from_image(image_url)
    entity_value = extract_entity(ocr_text, entity_name)  # Extract the specific entity
    
    # Append the results (group_id and prediction) to the list
    results.append({"group_id": group_id, "prediction": entity_value})
```
